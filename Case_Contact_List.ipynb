```python
import json


def save_agenda():
  with open('agenda.json', 'w') as file:
    json.dump(agenda, file)
  print("Saved!\na")

def load_agenda():
  try:
    with open('agenda.json', 'r') as file:
      agenda = json.load(file)
    print("Agenda loaded!")
    return agenda
  except:
    print("File not found")
    return []

# Function add contact
def add_contact(name, email, phone):
  contact = {
      'name': name,
      'email': email,
      'phone': phone,
  }
  agenda.append(contact)

# Function list contacts
def list_contacts():
  print('\n*** Contacts ***')
  for contact in agenda:
    print('- Name: %s' % (contact['name']))
    print('- Email: %s' % (contact['email']))
    print('- Phone: %s \n' % (contact['phone']))

# Function edit contact
def edit_contact(name):
  for contact in agenda:
    if contact['name'] == name:
      contact['email'] = input('email: ')
      contact['phone'] = input('Phone: ')
      print('Contact edited \n')
      break
  else:
    print('Contact not found \n')


def delete_contact(name):
  for contact in agenda:
    if contact['name'] == name:
      agenda.remove(contact)
      print('Contact deleted \n')
      break
  else:
    print('Contact not found \n')

# Main
agenda = load_agenda()

while True:
  print('*** Contact List ***')
  print('1 - Add contact')
  print('2 - List contacts')
  print('3 - Edit contact')
  print('4 - Delete contact')
  print('5 - Save')
  print('6 - Exit')
  option = input('Select an option: ')

# Colecting information
  if option == '1':
    name = input('Name: ')
    email = input('email: ')
    phone = input('Phone: ')
    add_contact(name, email, phone)
    print('\n Contact added: %s' %(name))

  elif option == '2':
    list_contacts()

  elif option == '3':
    name = input('Name: ')
    edit_contact(name)

  elif option == '4':
    name = input('Name: ')
    delete_contact(name)

  elif option == '5':
    save_agenda()

  elif option == '6':
    print('Exit')
    break

  else:
    print('Invalid option')
```
